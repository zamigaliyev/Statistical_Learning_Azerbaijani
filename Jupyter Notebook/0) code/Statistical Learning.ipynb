## Level of Measurements (Ölçmələr Səviyyəsi) 
* Levels of Measurements 4 ədəddir: nominal, ordinal, interval və ratio
* Nominal
    * Kateqoriyalar var.
    * Sıralanmazlar.
    * Məsələn Asiya ölkələri: Çin, Hindistan kimi
    * Ya da qadın/kişi kimi
* Ordinal:
    * Sıralana bilir amma aralarındakı əlaqə ədədi olaraq mənasızdır.
    * Məsələn bir anket düşünün şıklar "əsla, bəzən, ümumiyyətlə, hər zaman" olsun. Ümumiyyətlə bəzəndən daha ümumi bir ifadədir, lakin ədədi olaraq nə qədər dediyimiz zaman bu sualın cavabı yoxdur. Bu tərz məlumatlara ordinal data deyilir.
* Nominal və ordinal arasında sıralanma fərqi var.
* Interval:
    * Müəyyən bir miqyası var.
    * Natural zero starting point yoxdur. Məsələn, sıfırdan başlayır və ya temperatur üçün 10 dərəcədən başlayır deyə bilmərik.
    * Məsələn, otaq temperaturu 20 dərəcənin 10 dərəcədən isti olduğu kimi müqayisə edə bilərik.
* Ratio:
    * True zero point vardır. Məsələn, yaş, kilo kimi
    * Müəyyən bir miqyası var.
* Bəzi elm adamları interval və ratio növlərini eyni növ hesab edirlər.

## Central Tendency - Mərkəzi Tendensiya(Meyllənmə)
* Mean: ortalama 
* Median: bir listdə ortada olan rəqəm
* Mode: bir listdə ən çox tapılan rəqəm
* Mean vs Median:
    * Mean outlier'lardan təsirlənə bilər bu səbəblə median önəmlidir. Məsələn bir şirkətdə artım istənirsə, işçilərin orta məvacibi və ya əmək haqqının medianı varsa nə axtarmaq lazımdır?

```python
#********************************
import matplotlib.pyplot as plt
plt.style.use("ggplot")
import warnings
warnings.filterwarnings("ignore")
#********************************
import numpy as np
from scipy import stats

yas = [1,2,3,5,6,7,7,10,12,13]
# Mean
mean_yas = np.mean(yas)
print(mean_yas)
# Median
median_yas = np.median(yas)
print(median_yas)
# Mode
mode_yas = stats.mode(yas)
print(mode_yas)
```

### Mean VS Median
* Bəzən mean istifadə etmək əvəzinə median-dan istifadə daha məntiqli ola bilər.
* Məsələn bir şirkət düşünək bu şirkətdə 10 işçi var. Bu şirkətin müdiri əgər ortalama əmək haqqı 5 dəyərindən azdırsa hər işçinin əmək haqqısına artım edəcək.

```python
salary = [1,4,3,2,5,4,2,3,1,500]
print("Mean of salary: ",np.mean(salary))
```

* Ortalama əmək haqqı 52.5 çıxdı. Müdir bu ortalama səbəbindən işçilərə çox əmək haqqı verdiyini düşünəcək və bu səbəblə əmək haqqılarına artım etməyəcəkdir.
* Ancaq bildiyiniz kimi, bu ədalətli deyil. Çünki burada 500-ə yaxın bir əmək haqqı var ki, biz onu xarici(outlier) olaraq adlandıra bilərik.
* Median bu outlier dəyərlərin qarşısını alır.

```python
print("Median of salary: ",np.median(salary))
```

* Median dəyəri 3. Və bu dəyər əmək haqqıların həqiqi dəyərini əks etdirir. 5 dən də kiçik olduğu üçün müdir tərəfindən bu işçilər əmək haqqılarını artımla alacaqlar.

## Dispersion (Dağılma)
* Dispersion dağılma(yayılma,səpələnmə,genişlənmə) deməkdir. (spread out)
* Range: Bir listdə ən böyük və ən kiçik rəqəm arasındakı fərqdir
* Variance: 
    * varyans-Dəyişiklik. Listdəki bütün nöqtələr ortalamadan nə qədər uzaqlıqdadır 
    * ![Time](variance.jpg)
    * Bəzi düsturlarda N (N-1) əvəzinə ola bilər. Bu sample variance(N-1) və population variance(N) deyə iki fərqli anlayış olmasından qaynaqlanır, amma hal-hazırda çox da önəmli deyil.
* Standard deviation (std): 
    * Variance'ın kvadrat köküdür.
    * Bir paylamada dəyərlərin gələcəkdə görəcəyimiz paylamalardakı ortalamadan nə qədər olduğunu göstərmək üçün istifadə edəcəyimiz bir konsepsiyadır. Əslində çox istifadə edilən və önəmli bir konsepsiyadır.
    * ![Time](var_vs_std.jpg)

```python
# range
yas = [1,2,3,5,6,7,7,10,12,13]
print("Range: ", (np.max(yas)-np.min(yas)))

# variance
print("Variance: ", (np.var(yas)))
var = sum((yas - np.mean(yas))**2)/len(yas)
print("Variance with formula: ",var)

# std
print("Std: ", (np.std(yas)))
std = np.sqrt(sum((yas - np.mean(yas))**2)/len(yas))
print("Std with formula: ",std)
```

```python
import matplotlib.pyplot as plt
y = np.random.uniform(5,8,100)
x1 = np.random.uniform(10,20,100)
x2 = np.random.uniform(0,30,100)
plt.scatter(x1,y,color="black")
plt.scatter(x2,y,color="orange")
plt.xlim([-1,31])
plt.ylim([2,11])
plt.xlabel("x")
plt.ylabel("y")
print("X1 mean: {} and meadian: {}".format(np.mean(x1),np.median(x1)))
print("X2 mean: {} and meadian: {}".format(np.mean(x2),np.median(x2)))
```

## Quartiles
* What is quartile? -quartile nədir?
    * 1,4,5,6,8,9,11,12,13,14,15,16,17
    * Median ardıcıllığın ortasında olan ədəddir. Bu vəziyyətdə 11 olardı.
    * Kiçik(Lower) quartile (birinci quartile (Q1) (25%), ən kiçik rəqəm ilə median yəni orta hesabla 1 ilə 11 arasındakı mediandır ki, bu da 6-dir..
    * Böyük(Upper) quartile (üçüncü quartile (Q3) (75%), yuxarıdakı suala görə 14 olacaq olan median və ən çox sayda yəni 11 ilə 17 arasındakı medianı tapırsınız.
* IQR(inter quartile range) = Q3-Q1
* Outliers: Q1 - 1.5*IQR and Q3 + 1.5*IQR onun hüdudlarından kənarda olan dəyərlər

```python
# libraries
import pandas as pd
import seaborn as sns
plt.style.use("ggplot")
import warnings
warnings.filterwarnings("ignore")
# read data as pandas data frame
data = pd.read_csv("data.csv")
data = data.drop(['Unnamed: 32','id'],axis = 1)
```

```python
data.head()
```

```python
data_bening = data[data["diagnosis"] == "B"]
data_malignant = data[data["diagnosis"] == "M"]
desc = data_bening.radius_mean.describe()
Q1 = desc[4]
Q3 = desc[6]
IQR = Q3-Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
print("Anything outside this range is an outlier: (", lower_bound ,",", upper_bound,")")
data_bening[data_bening.radius_mean < lower_bound].radius_mean
print("Outliers: ",data_bening[(data_bening.radius_mean < lower_bound) | (data_bening.radius_mean > upper_bound)].radius_mean.values)
```

## Bivariate Data and Covariance (İki Dəyişənli Məlumat və Kovarianlıq)
* Bivariate
    * İki variable'ı qarşılaşdırır və correlation var mı yox mu ona baxır.

    * y oxu dependent variable-dır. Dependent çünki x-ə bağlıdır.

    * İki variable arasındakı əlaqə amma causality ilə qarşılaşdırmamaq lazımdır.
    * causality səbəbiyyət yəni iki variable arasında bir səbəb nəticə əlaqəsi var.

* Covariance

    * radius_mean and area_mean arasındakı covariance'a baxaq.

```python
np.cov(data.radius_mean,data.area_mean)
print("Covariance between radius mean and area mean: ",data.radius_mean.cov(data.area_mean))
print("Covariance between radius mean and fractal dimension se: ",data.radius_mean.cov(data.fractal_dimension_se))
fig, axs = plt.subplots(1, 2)
axs[0].scatter(data.radius_mean, data.area_mean)
axs[1].scatter(data.fractal_dimension_se, data.radius_mean)
plt.show()
```

## Pearson Correlation Coefficient (Pearson Nisbət Əmsalı)
* Pearson Correlation Coefficient: covariance'ı variable'ların standart deviation'ına bölürük
* Division of covariance by standart deviation of variables
* radius mean and area mean arasındakı pearson correlation əmsalına baxaq.
* Pearson Correlation Coefficient +1 və -1 dəyərləri arasında dəyişir.
    * +1 = positive linear correlation
    * -1 = negative linear correlation
    *  0 = linear correlation yoxdur 
* ![Time](pearson.jpg)

```python
p1 = data.loc[:,["area_mean","radius_mean"]].corr(method= "pearson")
p2 = data.radius_mean.cov(data.area_mean)/(data.radius_mean.std()*data.area_mean.std())
print('Pearson correlation: ')
print(p1)
print('Pearson correlation: ',p2)
```

```python
sns.jointplot(data.radius_mean,data.area_mean,kind="regg")
plt.show()
```

## Spearman Rank Coefficient (Spearman Dərəcə Əmsalı)
* Pearson correlation works well if the relationship between variables are linear and variables are roughly normal. But it is not robust, if there are outliers.

Yəni dəyişənlər arasındakı bağlantı xətti və dəyişənlər arasındakı əlaqə təqribən normaldırsa, Pearson korrelyasiyası yaxşı işləyir. Xarici məhsullar(Outliers) varsa,sağlam deyil.

* To compute spearman's correlation we need to compute rank of each value.

Yəni Spearman`nin nisbətini hesablamaq üçün hər bir dəyər dərəcəsini hesablamaq lazımdır.

```python
ranked_data = data.rank() 
spearman_corr = ranked_data.loc[:,["area_mean","radius_mean"]].corr(method= "pearson")
print("Spearman's correlation: ")
print(spearman_corr)
```

* Spearman's correlation is little higher than pearson correlation 

(Spearman'ın korrelyasiyası, pearson korrelyasiyasından bir qədər yüksəkdir)

    * If relationship between distributions are non linear, spearman's correlation tends to better estimate the strength of relationship 
    (Dağılmalar arasındakı əlaqə xətti deyilsə, spearman korrelyasiyası əlaqənin gücünü daha yaxşı qiymətləndirməyə meyllidir)

    * Pearson correlation can be affected by outliers. Spearman's correlation is more robust.
    (Pearson korrelyasiyasına outliers təsir edə bilər. Spearman'ın əlaqəsi daha möhkəmdir).

* One of the summary statistics.
-Kiçik şərh statistikalarından biridir.
* It describes size of an effect. It is simple way of quantifying the difference between two groups.
-Bir təsirin ölçüsünü təsvir edir. İki qrup arasındakı fərqli kəmiyyət hesablamaq üçün sadə bir üsuldur.
* In an other saying, effect size emphasises the size of the difference
-Başqa bir sözlə, təsir ölçüsü fərqin ölçüsünü vurğulayır.
* Use cohen effect size
-Koen effekti ölçüsündən istifadə edək.
* Cohen suggest that if d(effect size)= 0.2, it is small effect size, d = 0.5 medium effect size, d = 0.8 large effect size.
-Cohen təklif edir ki, d (təsir ölçüsü) = 0.2 olarsa, bu kiçik effekt ölçüsü, d = 0.5 orta təsir ölçüsü, d = 0.8 böyük effekt ölçüsüdür.
* lets compare size of the effect between bening radius mean and malignant radius mean
-size of the effect bening radius mean və malignant radius mean arasında müqayisə etməyə imkan verir
* Effect size is 1.9 that is too big and says that two groups are different from each other as we expect. Because our groups are bening radius mean and malignant radius mean that are different from each other
-Effekt ölçüsü 1.9 çox böyük olduğunu və iki qrupun gözlədiyimiz kimi bir-birindən fərqli olduğunu söylədi. Çünki qruplarımız bening radius mean və malignant radius mean-in bir-birindən fərqli olması deməkdir
* ![Time](efect.jpg)

```python
mean_diff = data_malignant.radius_mean.mean() - data_bening.radius_mean.mean()    # m1 - m2
var_bening = data_bening.radius_mean.var()
var_malignant = data_malignant.radius_mean.var()
var_pooled = (len(data_bening)*var_malignant +len(data_malignant)*var_bening ) / float(len(data_bening)+ len(data_malignant))
effect_size = mean_diff/np.sqrt(var_pooled)
print("Effect size: ",effect_size)
```

## Bayes Theorem (Bayes teoremi)
![Time](bayes.jpg)
* P(A|B) = B halı baş verdiyində A halının gerçəkləşmə ehtimalı
* P(A) = A halının gerçəkləşmə ehtimalı
* P(B|A) = A halı gerçəkləşdiyində B halının gerçəkləşmə ehtimalı
* P(B) = B halının gerçəkləşmə ehtimalı
* Bu düsturun necə çıxdığını görək
    * P(A|B) = P(A and B) / P(B)
    * P(B|A) = P(B and A) / P(A)
    * P(A and B) = P(B and A)
    * P(A|B).P(B)=P(B|A).P(A)
    * P(A|B) = P(B|A) . P(A) / P(B)
* Nümunə:
    * Bir məktəb var. Bu məktəbdə 50 nəfər şagird olsun. Bu 50 şagirddən 1 nəfəri x xəstəliyinə tutula bilər. X xəstəliyinə sahib xəstəlikli bir şagirdin testi %90 pozitiv, sağlam bir şagirdin testi isə %10 pozitiv nəticə verə bilər. Test nəticəsi pozitiv olan bir şagirdin həqiqətən x xəstəliyinə sahib olmaq ehtimalı nədir?
       * P(A) : Şagirdin xəstə olması ehtimalı = 1/50
       * P(B) : Testin pozitiv çıxması ehtimalı = 1/50 * 0.90 + 49/50 * 0.10 = 0.116
       * P(A|B) : Pozitiv çıxan testin xəstəlik çıxma ehtimalı nədir ? 
       * P(B|A) : x xəstəliyinə sahib şagirdin testinin pozitiv çıxma ehtimalı = 0.9
       * P(A|B)=P(B|A)*P(A)/P(B) => (0.9 * 1/50) / (0İndi bu formulun necə çıxdığına baxaq.116) = %15

## Gaussian (Normal) Distributions
* Continuous Probability Distributions "Probability density functions (PDF)" olaraq da adlandırılır
* Gaussian ayrıca Bell Shaped Distribution olaraq da adlandırılır.
* Gasussian distribution'ın vacib olmasının bir səbəbi real həyatda bir çox nümunəsi olmasıdır.
    * IQ
    * Boy və çəki
* Simmetrikdir. Bir mean dəyəri ətrafında müəyyən bir standart deviation'a görə paylanma göstərirlər.
* Instead of making formal definition of gaussian distribution, I want to explain it with an example.
* The classic example is gaussian is IQ score.
    * In the world lets say average IQ is 110.
    * There are few people that are super intelligent and their IQs are higher than 110. It can be 140 or 150 but it is rare.
    * Also there are few people that have low intelligent and their IQ is lower than 110. It can be 40 or 50 but it is rare.
    * From these information we can say that mean of IQ is 110. And lets say standart deviation is 20.
    * Mean and standart deviation is parameters of normal distribution.
    * Lets create 100000 sample and visualize it with histogram.

```python
# parameters of normal distribution
mu, sigma = 110, 20  # mean and standard deviation
s = np.random.normal(mu, sigma, 100000)
print("mean: ", np.mean(s))
print("standart deviation: ", np.std(s))
# visualize with histogram
plt.figure(figsize = (10,7))
plt.hist(s, 100, normed=False)
plt.ylabel("frequency")
plt.xlabel("IQ")
plt.title("Histogram of IQ")
plt.show()
```

* As it can be seen from histogram most of the people are cumulated near to 110 that is mean of our normal distribution
* However what is the "most" I mentioned at previous sentence? What if I want to know what percentage of people should have an IQ score between 80 and 140?
* We will use z-score the answer this question. 
    * z = (x - mean)/std 
    * z1 = (80-110)/20 = -1.5
    * z2 = (140-110)/20 = 1.5
    * Distance between mean and 80 is 1.5std and distance between mean and 140 is 1.5std.
    * If you look at z table, you will see that 1.5std correspond to 0.4332
    * ![Time](z_score.jpg)  
    * Lets calculate it with 2 because 1 from 80 to mean and other from mean to 140
    * 0.4332 * 2 = 0.8664
    * 86.64 % of people has an IQ between 80 and 140.
    * ![Time](hist_iq.jpg)

## Hypothesis Testing
* Hipotez testlər statistik metodların real problemlərə tətbiq edilməsidir.
* **Null Hypothesis** məsələn, dünya düzdür. Bu fərziyyə ya rədd edilir, ya da rədd edilə bilmir.
    * reject: rədd etmək
    * fail to reject: rədd etməmək
* Əgər null hypothesis reject edilirsə. Bir ədəd **alternate hypothesis** yaradırıq. Məsələn yer kürəsi dairəvidir.  
* Həqiqi yer kürəsi problemini hipotezə necə çevirdiyimizə baxaq..
    * İlk əvvəl bir null hypothesis yaradırıq, başlanğıcda bu null hypothesis doğru (true) olaraq qəbul edirik və sonra çürütməyə(reject) çalışırıq. Əgər data null hypothesis'i reject edərsə alternative hypothesis'lər yaratmağa başlayırıq. 
    * Null hypothesis "Bakı'da insanların boy ortalaması 2 metr" olsun.
    * Null hypothesis: average height = 2 m
    * Alternate hypothesis "Bakı'da insanların boy ortalaması 2 metr deyil"
    * Alternate hypothesis: average height != 2 m
    * Başlanğıcda null hypothesis doğru olaraq qəbul edirik demişdik. İndi ilk əvvəl datamızı toplayırıq.
    * Sonra test statistikalarını (test statistic = z value) istifadə edərək p-value'nu tapırıq.
    * p-value'nu 0.05 ilə qarşılaşdırırıq. 0.05 = level of significance
    * Əgər p-value < 0.05(level of significance) (confidence level 95%)
    * Null hypothesis'i reject edirik. If p-value is low, we conclude that null hypothesis is unlikely to be true
    * Əgər p-value > 0.05(level of significance) fail to reject edirik.

## Hypothesis Testing Real-World Nümunəsi 1
* Real World example!
* Bir avtomobil şirkəti dayanacaq sensorları düzəldir.
* Bu sensor ortalama 3 milli saniyədə bir sample alır və bu zamanın xəta payı 0.6 ms-dir. Yəni 2.4-3.6 ms aralarında ölçü ala bilər
    * **mean = 3 ms**   ve **sigma = 0.6 ms**
* Avtomobil şirkəti sahibi bu saatı 3 ms azaltmağı söyləyir və bunun üçün mühəndislər işə götürür
* Ayrıca, sahibi bu sensorun daha qısa müddətdə, 99% nisbətində ölçmə aparmasını istəyir. Beləliklə, güvən səviyyəsi (confidence level)= 99% => ** a = 0.01 **
* Mühəndislər bu sensoru inkişaf etdirdikdən sonra 50 dəfə işləyib sınaqdan keçirirlər. Nəticə budur ki, bu sensor indi orta hesabla 2.8 ms çəkə bilər. Beləliklə ** n = 50 ** və ** yeni_mean = 2.8 ms **
* Sual olunur, yeni sensorla alınan bu ölçmələr statistik olaraq əvvəlki sensordan daha yaxşıdır?
* Əvvəlcə burada iki ehtimal var, bəli mühəndislər yaxşı iş gördülər və sensor daha sürətli. İkincisi, yox, ilk sensorun artıq 0,6 həddi var, buna görə 2.8 ms yeni nəticə təsadüfən çıxa bilər.
* Bu sualın cavabını bizə **hypothesis testing** verəcək.
* İlk olaraq null hypothesis təyin edilir. Sonra alternative hypothesis
    * Null hypothesis mean >= 3 (burada null hypothesisi çürütməyə çalışacağıq)
    * Null hipotezi müşahidə edilən məlumatlardan istifadə edərək yalan olduğu sübut edilə bilən izahatlardır.
    * Alternate hypothesis mean < 3

* Test type: left tail (alternative hypothesis'e baxılır)

* P = 0.0094
* Əgər p-value > 0.01(level of significance) fail to reject edirik
* 0.0094 < 0.01 bu səbəbdən **reject to Null Hypothesis**
* Yəni null hypothesis yanlışdır. Sensorumuz sürətlənib. Mühəndislər uğurlu oldular.

## Chi-Square Analysis Nümunə
* Bir internet kafemiz var.
* Burada 7 ədəd bilgisayar var.
* Bu 7 bilgisayarın çökmə ehtimalı birbiri ilə eynidir.
* Yəni expected = failure rate should be same for all computers
* Bilgisayarlar birbiri ilə independentdir. 
* Observations: 1(5), 2(7), 3(9), 4(4), 5(1), 6(10), 7(6)
* Null hypothesis = observation dəyərlərinin bu şəkildə çıxması statistik olarak %95 ehtimal ilə məntiqlidir(uyğundur,doğrudur).
* Toplam çökmə 42.
* Expected value = 42 / 7 = 6
* degrees of freedom = 7 - 1 = 6
* 12.5 > 9.3 fail to reject null hypothesis

```python
observation = np.array([5,7,9,4,1,10,6])
print("total: ",np.sum(observation))
expected = np.sum(observation)/ len(observation)
print("expected: ",expected)
chi_value = np.sum(((observation - expected)**2)/expected)
print("chi_value: ",chi_value)
```
